## 산점도

* 간단한 예제

```python
# 데이터 생성
np.random.seed(3)

x = 10 + np.random.normal(0, 1, 100)
y = 10 + np.random.normal(0, 1, len(x))

# size, color 설정
size = np.random.uniform(5, 30, len(x))
color = np.random.uniform(5, 30, len(x))

# 그래프 그리기
fig, ax = plt.subplots()

ax.scatter(x, y, s=size, c=color)
ax.set(xlim=(5, 15), xticks=np.arange(5, 15),
       ylim=(5, 15), yticks=np.arange(5, 15))

plt.grid()
plt.show()
```

* 실데이터 확인하기

```python
# 방송매체와 Sales와의 관계
fig, ax = plt.subplots()

# 산점도 그리기
ax.scatter(ad.TV, ad.Sales, s=5, c='blue')
ax.scatter(ad.Radio, ad.Sales, s=5, c='green')
ax.scatter(ad.Newspaper, ad.Sales, s=5, c='red')

# 그래프 옵션 설정하기
ax.set(xlabel='TV Radio Newspaper', ylabel='Sales')

plt.grid()
plt.show()
```

```python
# 3D로 만들 수 있을 것 같은데..?
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 15))

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Create the scatter plot
ax.scatter(ad.TV, ad.Radio, ad.Sales, c='r', s=5)
ax.scatter(ad.TV, ad.Newspaper, ad.Sales, c='b', s=5)
ax.scatter(ad.Radio, ad.Newspaper, ad.Sales, c='g', s=5)

# Set labels
ax.set_xlabel('TV, TV, Radio')
ax.set_ylabel('Radio, Newspaper, Newspaper')
ax.set_zlabel('Sales')

# Set a title
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```

# PairGrid

* pairplot의 불편함을 없애려고 나온 것
* 대각선을 기준으로 위, 아래에 대해 각각의 시각화 도구를 적용할 수 있음.
* 숫자형 - 숫자형의 관계를 기본으로 그린다. (숫자형-범주형도 그릴 수 있음.)

FacetGrid

* 다양한 범주형 값을 가지는 데이터를 시각화하는 것에 좋은 방법
* 행, 열 방향으로 서로 다른 조건을 적용해 여러 개의 서브플롯을 만든다.
* 각 서브 플롯에 적용할 그래프 종류를 map() 메서드를 이용해 그리드 객체에 전달한다.

```python
g = sns.PairGrid(ad)

# 대각선에는 histplot을 나오게 한다.
g.map_diag(sns.histplot)
# 대각선이 아닌 것 >> 산점도 & 회귀선으로 표현하기
g.map_offdiag(sns.regplot)

plt.tight_layout()
plt.show()
```

# EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from matplotlib import font_manager, rc, cm
import matplotlib as mpl

# 마이너스 깨짐 현상 해결
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothic'
```

## Credit data

```python
# 기본 path 설정
path = '../datasets/ML_data/'

raw = pd.read_csv(path+'Credit.csv')
credit = raw.copy()
print(credit.shape)
credit.head(2)
```

```python
credit.info()
```

```python
credit.describe(include='all')
```

```python
# 범주형 : Cards, Education
credit.Education.unique()
```

```python
# 범주형 빼고 시각화 확인하기
g = sns.PairGrid(credit[['Income', 'Limit', 'Rating', 'Age', 'Balance']])

g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

plt.tight_layout()
plt.show()
```

```python
# Balance와 Income, Student의 관계성 확인하기 - scatterplot
fig, ax = plt.subplots()

# 그래프 그리기
sns.scatterplot(data=credit, x=credit.Income, y=credit.Balance, hue='Student')
ax.set(xlabel='Income', ylabel='Balance')

plt.grid()
plt.show()
```

```python
# Boxplot
fig, ax = plt.subplots()

sns.boxplot(x=credit.Student, y=credit.Balance, data=credit, hue='Student')

plt.grid()
plt.show()
```
