# LinearRegression

단순 선형 회귀 - y = ax + b

* **Error의 척도(MSE, MAE 등 - Cost Function)를 낮추기 위해** a, b의 값을 최적화하는 방법을 이용
* Error = |예측값 - 실제값|
* 입력 변수의 개수가 늘어나면 Multinomial Linear Regression이 되고
* Sigmoid 함수 안으로 들어가면 Logistic Regression이 된다.

Linear Regression의 전제 (보통, 정규성, 선형성, 등분산성을 확인한 후, 회귀 분석을 진행한다.)

* Normality 정규성 : 잔차의 정규성
    1. QQ plot를 통한 시각화 자료 확인
        * 데이터가 직선 상에 있음 >> 정규성을 따른다.
    2. 히스토그램을 통한 시각화 자료 확인
        * 히스토그램 막대가 정규분포 곡선 상에 있음 >> 정규성을 따른다.
    3. 정량적 검정
        1) Normality test : Shapiro-Wilk test, Kolmogorov-Smirnov test
            * Shapiro-Wilk : 2000개까지 / Kolmogorow-Smirnov : 2000개 이상
        2) Ordinal logistic regression의 proportional odds assumption test : Score test
        3) Cox regression의 proportional hazard assumption test : Schoenfeld residual test
        * p-value > 0.05일 때 >> 정규성을 위반한다고 볼 수는 없다.
        

* Linearity 선형성 : 두 변수 (독립변수-종속변수) 간의 선형 관계
* Homoscedasticity 잔차의 등분산성 : 잔차의 분산은 종속변수의 값에 따라 달라지지 않음.
* Independence 독립성 : 관측 간의 독립

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 선형 회귀 관련 사이킷런
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

import statsmodels.api as sm
```

* 간단한 예제

```python
# 데이터 준비
X = np.array([[0], [1], [2]])
y = np.array([0, 1.2, 1.6])

# model 변수 저장
model = linear_model.LinearRegression()

# model 훈련하기
model.fit(X, y)

# train 예측
y_pred = model.predict(X)

# test 예측
test_pred = model.predict([[1.5]])
test_pred
```

```python
# coefficient
print('coef :', model.coef_)
```

```python
# 시각화
# 실제값
plt.scatter(X, y, color='red')
# 예측값
plt.scatter(1.5, test_pred, color='green')

# 회귀선 그리기
plt.plot(X, y_pred, color='black', linewidth=1)
plt.xlabel('X')
plt.ylabel('y')

plt.grid()
plt.show()
```

```python
path = '../datasets/ML_data/'

raw = pd.read_csv(path+'Advertising.csv', index_col=0)
ad = raw.copy()
print(ad.shape)
ad.head(2)
```

Error의 기준 (Cost Function)

1. MSE
* Error가 정규분포의 특성을 보일 때 적합하다.
* Outlier의 영향을 상대적으로 많이 받는다.
* ML에서 대용량의 데이터를 다룰 때, Gradient Descent 방식을 이용하는 경우가 많음. 이때, 값이 줄어들면서 미분값도 줄어들어 수렴할 수 있다.

2. MAE
* Outlier의 영향을 많이 받지 않는다.
* 기울기가 일정하기 때문에, 최저점으로 수렴이 안 될 가능성이 있음.

3. Huber, Log-cosh
* MAE의 Outlier에 대해 Robust한 특징과 MSE의 Gradient가 최적점 근처에서 줄어듦으로써 더 정말한 값을 찾을 수 있도록 하는 장점을 결합.

4. Quantile
* 예측값의 분산은 항상 일정하다는 것을 바꾸지 않고 quantile을 이용해 예측 구간 표현 가능

```python
# 선형회귀모델 적용하기
# train/test data 분리하기 >> 과적합 방지
train = ad[:-20]
test = ad[-20:]

# train feature / label
X_train = train.iloc[:, :-1]
y_train = train[['Sales']]

# test feature / label
X_test = test.iloc[:, :-1]
y_test = test[['Sales']]
```

* 사이킷런 사용

```python
# TV
# 모델링 - 선형회귀
model = linear_model.LinearRegression()

# 모델 학습
model.fit(X_train[['TV']], y_train)

# train data 예측하기
y_train_pred = model.predict(X_train[['TV']])

# test data 예측하기
y_test_pred = model.predict(X_test[['TV']])

# 기울기
print('coef :', model.coef_)
print('y절편 :', model.intercept_)
# MSE 확인
print("train data's MSE :", round(mean_squared_error(y_train, y_train_pred), 3))
print("test data's MSE :", round(mean_squared_error(y_test, y_test_pred), 3))
# 결정계수 확인
print("train data's R^2 score :", round(r2_score(y_train, y_train_pred), 3))
print("test data's R^2 score :", round(r2_score(y_test, y_test_pred), 3))
```

```python
y_test_pred
```

```python
# 시각화하기
# 산점도
plt.scatter(X_train.TV, y_train.Sales, color='gray')
plt.scatter(X_test.TV, y_test.Sales, color='red')
# 회귀선 그리기
plt.plot(X_train.TV, y_train_pred, color='green', linewidth=2)
# 예측값 확인하기
plt.scatter(X_test.TV, y_test_pred, color='blue', marker='^')

# 그래프 꾸미기
plt.title('Linear Regression Result')
plt.xlabel('TV')
plt.ylabel('Sales')

plt.grid()
plt.show()
```

* statsmodels 사용

```python
# OLS 사용
# train / test data에 x0 컬럼 만들어주기
# 왜냐면, OLS 함수는 데이터 내에 intercept하는 feature가 필요하다.
# train
sm_X_train = X_train.copy()
sm_X_train['x0'] = 1

# test
sm_X_test = X_test.copy()
sm_X_test['x0'] = 1

# 모델 생성하고 훈련하기
model = sm.OLS(y_train, sm_X_train[['x0', 'TV']]).fit()

# train data로 예측하기
y_train_pred = model.predict(sm_X_train[['x0', 'TV']])

# test data로 예측하기
y_test_pred = model.predict(sm_X_test[['x0', 'TV']])

# 결과 확인하기
model.summary()
```

# Resampling 재표본추출 방법

Validation set

* train data로부터 **반복적으로** 표본을 추출
* --> 모델에 대한 새로운 정보를 얻기 위해 모델을 새로 추출한 sample에 적합 (모델을 여러 번 적합(fitting))
* 대표적인 방법 : **cross-validation(교차 검증)**, **Boostrap**
* 머신러닝의 응용 어플리케이션에 중요하게 사용된다.
    * model assessment : 모델의 성능을 평가하는 과정에 사용
    * model selection : 모델의 적절한 유연성을 찾는 과정에 사용
 
* 모델을 train data로 훈련한다고 가정한다. 하지만 과적합될 수 있기 때문에 test data로 성능평가를 해야 한다.
    * 성능평가 지표 : MSE
 
* prediction error 추정방법
    1. 많은 양의 test dat 확보
    2. 수학적 조정
    * train error로 test error 추정할 수 있도록 조정하기
    * ex) Cp statistic, AIC, BIC 등
  
    3. train data의 일부를 test에 사용
 
* Validation set 방법
    * training data를 임의(random)로 두 부분으로 구분
        * training set / Validation set (**hold-out** set)
    * Training, validation 방법
    1. 모델을 training set에 fitting
    2. training된 모델을 validation set으로 평가
    * Validation-set error는 test error의 추정값
        * 주로 MSE, 오분류율(misclassification rate)로 평가함.
     
K-fold cross-validation

* LOOCV (leave-one out cross-validation)
    * validation set : data point 1개
    * training set : validation set을 제외한 나머지
    * test error 추정
        * 하나의 validation data point에 의존하기 때문에 변동성이 크다.
        * error 추정과정을 여러번 반복해 추정한다.
    * validation set 방법과 비교 (장점)
        * n-1개의 data point를 training에 사용하기 때문에 편향이 낮다. (= 실제 test error보다 지나치네 높게 추정하지 않는다.)
        * 여러 번 시행해도 같은 결과를 얻는다.
     
* **K-fold cross-validation**
1. training data를 임의로 k부분으로 나눈다.
2. k번째 부분은 빼고 k-1개의 부분으로 모델 적합 후, k번째 부분으로 성능평가한다.

* LOOCV와 K-fold 비교
1. Bias-variance trade-off 관점
    * k-fold CV는 LOOCV와 validation set 방법의 중간정도의 bias를 가진다.
    * LOOCV는 n개의 모델이 모두 높은 상관관계를 가져 test error 추정값의 변동성이 크다.
2. test error 추정의 정확성 측면
    * 두 방법 모두 비슷한 경향을 보이지만 모두 test error를 과도/과소 추정한다.
    * flexibility를 정하는 데 좋은 성능을 보인다.
  
* 분류 문제의 cross validation
    * error를 측정하는 방법만 제외하고 회귀문제와 동일

```python
import numpy as np

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import KFold
from sklearn.utils import resample

from sklearn.utils import shuffle

from sklearn.datasets import load_iris
```

* 간단한 예제

```python
# 데이터 준비
X = ['a', 'b', 'c', 'd', 'e', 'f']

# 모델 생성
kfold = KFold(n_splits=3) # 3개로 나누기

for train, test in kfold.split(X):
    print(f'{train} {test}')
    print(X[train[0]], X[train[1]])
    print(X[test[0]], X[test[1]])
```

* 실데이터 분석

```python
# 데이터 준비
X, y = load_iris(return_X_y=True)

# 데이터 모델링
# 5 fold cross validation
kfold = KFold(n_splits=5)
idx = 1

for train, val in kfold.split(X):
    # 몇 번째 fold
    print('='*10, 'Fold #', idx, '='*10)
    # LDA 적용하기
    model = LinearDiscriminantAnalysis().fit(X[train], y[train])
    
    ## 예측하기
    # 원래의 validation 정답
    print('True Label :', y[val])
    # lda로 예측한 validation의 정답
    print('Predicted Label :', model.predict(X[val]))

    ## 정확도 확인하기
    print('Accuracy :', model.score(X[val], y[val]) * 100, '%')

    idx += 1
```

# Model Selection

선형모델 (linear_model)

* 실용적 (성능), 개념적 (X와 Y의 관계 해석) 측면에서 매우 유용하다.
* 최소제곱법의 **대안적인 적합 방법**을 통해
    * 예측 정확도 (prediction accuracy) : 모델의 variance를 조정해 예측 성능을 향상시킨다.
    * 모델 해석 (model interpretability) : 관련 없는 feature를 제외해서 조금 더 해석이 잘 되는 모델을 생성한다.
* 3가지의 대안적인 방법
    1) Subset Selection (부분 집합 선택)
    * p개의 predictor 중 response와 관련된 것이라고 생각되는 predictor 식별
    * 식별된 p개보다 적은 수의 predictor만을 least squares 방법으로 fitting
    
    2) shrinkage (수축)
    * p개의 predictor로 fitting하되, **coefficient의 추정값이 0**으로 작아지도록 fitting.
    * **정규화**라고도 불림.
    * **모델의 variance를 줄이고 변수를 선택**하는 효과를 가짐.

    3) dimension reduction (차원 축소)
    * p개의 predictor를 **M차원의 subspace**에 투사하는 방법 (M < p)
    * M개의 linear combination을 만들어내 선형회귀의 predictor로 사용한다.
    * 비지도 학습의 방법 중 하나이다.
 
1. Subset Selection
1) Best subset selection
    * 알고리즘 : 모든 predictor 조합 중 최고 성능을 내는 조합을 선택.
2) Stepwise selection
    * best subset은 **계산량이 많아** p가 매우 클 경우 문제가 됨. ==> 2^p개의 조합 존재
    * p가 클 경우, 통계적인 문제를 가질 수도 있음.
        * Search space(검색 영역)이 넓을 경우, training data에 **과적합**되는 모델을 찾게 될 확률이 높아진다.
    * **계산량과 과적합 문제에 대한 대안**으로 stepwise 방법을 사용한다.
        * best subset보다 훨씬 적은 수의 모델을 탐색한다.
3) Forward stepwise selection
    * predictor가 없는 모델에서 시작해 모든 predictor가 모델에 사용될 때까지 한번에 하나씩 predictor를 추가해 나가는 방법
    * **계산량** 측면에서 best subset 대비 장점이 있다.
4) Backward stepwise selection
    * 모든 predictor가 포함된 모델에서 시작해 **가장 유용하지 않은 predictor를 한번에 하나씩 제거**해나가는 방법
  
2. 최적 모델 선정
* 모든 predictor를 포함하는 모델이 항상 **최고 RSS와 최대 R^2**를 가진다.
* 최적의 모델은 낮은 train error가 아닌 **낮은 test error**를 가진 모델
* ==> predictor의 개수가 서로 다른 모델 중 best 모델을 선정하는 경우, 단순히 RSS와 R^2를 비교하는 것은 부적절하다.
    1) 간접적인 방법
    * **training error의 조정**을 통해 test error를 간접적으로 추정 (과적합때문에 발생하는 편향을 고려하는 방법)
    * Cp, AIC, BIC, adjusted R^2 등이 있고,  그 계산이 매우 간단함.
        1. Mallow's Cp
        2. AIC : 최대우도를 이용한 기준 (Cp와 비례)
        3. BIC
        * 가장 작은 BIC값을 가진 모델 선택
        * Cp의 2𝑑𝜎Ƹ2항을 log(𝑛)𝑑𝜎Ƹ2항으로 대체한 형태
            * variable이 많은 경우, Cp보다 더 작은 모델을 선정하는 경향을 보인다.
        
        4. Adjusted R^2
        * 값이 클수록 작은 error를 가진 모델
        * 불필요한 variable을 포함하는 것을 막을 수 있게 해주는 보정을 한다.

    2) 직접적인 방법
    * **validation set 또는 cross-validation 방법**을 이용해 직접적으로 test error를 추정
    * **모델을 여러 번 fitting**해야하기 때문에 계산량이 많다.
    * 모델의 유연성을 정의하기 어렵거나 𝜎2를 추정하기 어려운 경우에 사용 가능

3. Shrinkage 방법
* Ridge regression과 Lasso (overfitting 방지용)
    * coefficient를 제한하거나 정규화하는 방법을 사용 (coefficient의 추정값을 0으로 축소시키는 것을 의미함.) ==> coefficient의 추정값의 variance를 상당히 줄일 수 있다.
    1) Ridge regression
    * L2 penalty에서 beta 1 ~ beta p가 0에 가까울 때, shrinkage penalty가 작아진다는 차이가 있음.
    * ==> coefficient의 추정값이 0으로 축소되는 효과를 가진다.
    * tuning parameter 𝜆 : 두 항의 상대적인 효과를 조절하는 역할
    * ==> 최적의 𝜆를 정하는 것이 매우 중요하고, 주로 cross-validation을 통해 정한다.
    * Predictor scaling
    
    2) Lasso
    * ridge regression의 단점 : 모든 predictor p개를 포함.
    * L1 penalty
    * 𝜆가 충분히 크면 coefficient를 정확히 0으로 만든다. ==> 적절한 𝜆를 정하는 것이 매우 중요함.
* 비교
    * response와 관련되어 있는 predictor의 개수가 적은 경우, lasso가 상대적으로 더 좋은 성능을 낼 것으로 기대할 수 있음.
    * 하지만, 실제 data에서는 response와 관련되어 있는 predictor가 어떤 것인지 몇 개인지 알 수 없기 때문에, **cross-validation**을 사용해서 두 방법 중 어떤 방법이 특정 data에 더 적합한지 결정해야 한다.

*  실데이터 분석

```python
path = '../datasets/ML_data/'

raw = pd.read_csv(path+'Advertising.csv', index_col=0)
ad = raw.copy()
print(ad.shape)
ad.head(2)
```

```python
# ad 데이터에 Lasso 적용하기 --> 최적의 lambda 선정
n_fold = 5

for t_param in (0.001, 0.01, 0.1, 1, 10, 100):
    print('='*15, 'Lambda :', t_param, '='*15)

    # 모델 생성
    kfold = KFold(n_splits=n_fold)
    idx = 1
    sum_val_mse = 0

    for train, val in kfold.split(ad):
        print('Fold #', idx)
        # train data의 feature, label
        X_train = ad.iloc[train][['TV', 'Radio', 'Newspaper']]
        y_train = ad.iloc[train][['Sales']]

        # validation data의 feature, label
        X_val = ad.iloc[val][['TV', 'Radio', 'Newspaper']]
        y_val = ad.iloc[val][['Sales']]

        # Lasso 적용하기
        model = linear_model.Lasso(alpha=t_param)

        # fitting
        model.fit(X_train, y_train)

        # coefficient
        print('coef :', model.coef_)

        # validation set 예측
        y_val_pred = model.predict(X_val)
        
        # validation MSE
        val_MSE = mean_squared_error(y_val, y_val_pred)
        print("validation's MSE :", val_MSE)

        # val_MSE 합하기
        sum_val_mse = val_MSE
        print('-'*50)
        idx += 1

    print('>>> Average Validation MSE :', round(sum_val_mse/n_fold, 3))
    print()
```

# 의사결정나무

**Regression Tree**

트리 기반 방법

* predictor space : 여러 대의 단순하고 작은 공간으로 **계층화** 또는 **나누는** 방법
* decision tree의 장단점
    * 장점 : 단순해서 해석하기 쉬움
    * 단점 : 보통 다른 지도학습 방법보다 성능이 좋지 못함.
    * ==> 여러 개의 트리를 만들어 예측 성능을 높이는 방식인 *bagging, random forest, boosting* 방법을 사용한다. (해석력은 떨어짐)
 
트리 생성 과정

* predictor space를 J개의 겹치지 않는 공간으로 구분한다.
* 각각의 데이터는 R 중 하나의 공간에 떨어지고 같은 공간에 있는 데이터들은 모두 같은 예측값을 가진다. (해당 공간의 training data의 response의 평균)
* 트리 생성 목표 : RSS를 최소화하는 box R1, R2, ..., Rj를 찾는 것
* ==> revursive binary splitting (반복적으로 2개의 공간으로 구분) 방법을 사용함.
* top-down 트리의 공간을 계속해서 분할해 매번 새로운 공간을 생성한다.
* 최적의 분할법 : RSS를 최소화하는 cutpoint를 찾는다.
* 중단 규칙을 만족할 때까지 공간 분할을 반복 수행한다.

트리 과적합 (overfitting) 방지

* J가 클 (= 트리가 깊은) 경우 (잘게 쪼갬) : 과적합 가능
* J가 작을 경우 : variance가 낮고 해석력이 좋아진다. 하지만, bias는 커진다.
* 대안 : 트리 생성 과정에서 트리가 너무 깊어지지 않도록 조절한다.
    * 공간을 분할할 때 발생하는 RSS 감소가 특정 threshold보다 클 때까지만 트리 생성
* 단점 : 향후 있을 큰 RSS의 감소를 만드는 공간 분할을 사전에 차단할 수 있다.

트리 가지치기 (Pruning a tree)

* 과적합을 막기 위한 방법

**Classification Tree**

* 한 공간에 가장 많이 포함되어 있는 training data의 클래스로 예측
* RSS 대신 classification error rate를 사용한다.
    * 한 공간에서 단순히 가장 많은 클래스에 포함되지 않는 training data의 비율
 
Classification Tree 생성 기준

* Gini index : K개 클래스별 data의 비율에 관련된 값을 측정하는 기준
    * node purity (노드의 순도)를 나타내는 값에 해당한다.
    * 지니계수가 낮을수록 노드의 순도가 높다.
* Cross-entropy : Gini index의 대안

트리의 장단점

* 장점
    * 설명하기 쉽다 (분기가 단순할 경우, 선형회귀보다 더 직관적이다.)
    * decision tree가 사람의 의사결정 방법과 유사하다는 믿음이 있다.
    * 그림으로 설명할 수 있어 비전문가도 이해하기 쉽다.
    * dummy variable 생성 없이 질적 변수를 사용할 수 있다.
* 단점
    * 다른 머신러닝 방법에 비해 성능이 낮다.
    * ==> 여러 개의 decision tree를 생성해 합치는 (aggregating) 방법을 통해 성능을 크게 향상시킬 수 있다.

**Bagging (Bootstrap aggregation)**
* 머신러닝 방법의 변동성을 줄이기 위한 일반적인 방법 --> decision tree 방법에 특히 유용해 많이 적용된다.
* Bagging
    1) B개의 bootstrap training data를 생성
    2) x의 response 𝑓መ∗𝑏를 예측하기 위해 b번째 training data로 fitting
    3) 모든 bootstrap training data에 대해서 fitting 후 평균을 구한다. (Regression)
    4) majority vote (가장 많이 선택된 클래스로 예측) 방법 사용 (Classification)
* Out of Bag Error Estimation (OOB)
    * Bootstrap은 중복을 허용하기 때문에 하나의 bootstrap training data에서 평균적으로 본래 데이터의 2/3가 샘플링된다.
    * fitting에 사용되지 않은 1/3의 데이터를 OOB이라고 한다.
    * i번째 데이터가 OOD인 경우의 decision tree에서 i번째 데이터의 response를 구하고 이들을 평균해 test error의 추정에 사용된다.
    * ==> 이 추정값은 B가 충분히 크면 LOOCV와 같다.
 
**Random forests**
* Bootstrap에 의해 생성된 트리들의 관계성을 낮춰 성능을 향상하는 방법
    * 관계성이 낮은 트리들을 평균해 variance를 낮춘다.
* 트리를 split할 때, p개의 predictor를 고려하는 것이 아니라 임의로 선택된 m개의 predictor만을 사용한다.
* 매 분기마다 새롭게 m개의 predictor를 선택

**Boosting**
* 트리를 이전에 학습한 트리로부터 순차적으로 천천히 학습시키는 방식
    * Bootstrap 샘플링을 하지 않고 본래 데이터에서 수정된 데이터를 이용해 학습한다.
* 학습 방식
    1) 아무 정보도 없는 트리 생성
    2) 트리를 response에 fitting하는 것이 아닌 residual에 fitting한다.
    3) residual에 적합해서 만든 트리를 기존 트리에 업데이트하고 residual도 업데이트 한다.
    4) 2, 3의 과정을 B회 반복한 후 최종 트리를 얻는다.
* 알고리즘 : 작은 트리 (소수의 terminal node)를 더해 나가는 방법
* tuning parameter
    1) 트리의 수 B : B가 크면 과적합될 수 있기 때문에 cross-validation으로 정한다.
    2) Shrinkage parameter 𝝀 : 작은 양수의 값으로 boosting이 얼마나 빠르게 학습되는지를 정한다. 보통 0.01 또는 0.001 정도로 정하고 값이 작을수록 더 큰 B가 필요하다.
    3) 분기의 수 d : 1인 경우에도 좋은 성능을 낸다.
 
**앙상블**
* 여러 개의 classifier로 결과를 생성 후 종합해서 majority vote하는 것처럼 여러 개의 모델을 활용해 성능을 높이는 방법

**Maximal Margin Classifier**

Binary Classification (two-class)
* 직관적인 분류 방법 : feature space에서 공간을 둘로 나누는 평면을 찾는다.
    * 그러한 평면을 찾을 수 없다면?
    * --> 공간을 구분한다는 개념을 느슨하게 적용한다.
    * --> 구분이 가능하도록 feature space 를 확장한다.

Hyperplane(초평면)
* p차원 공간에서 p-1 차원의 부분 공간 (subspace)
* separating hyperplane : training data의 class label을 완벽하게 분리해내는 hyperplane
* f(x)의 절댓값이 크면 조금 더 클래스의 분류에 대한 신뢰도가 높다.

Maximal Margin Classifier
* 모든 separating hyperplane 가운데 두 class간의 margin을 가장 크게 하는 separating hyperplane이 존재한다.
* Support vector : Maximal Margin Classifier를 정의하기 위해 지원하는 data
* **최적화 문제 (optimization problem)**

Noisy 데이터
* Linear boundary로 구분할 수 없는 데이터

**Support Vector Classifier**
* linear kernel을 활용
* Soft margin : 모든 데이터를 완벽히 구분해내는 margin이 아닌 일부 training data에 대해서는 잘못된 클래스 구분을 허용하는 margin
* Support Vector Classifier(Soft Margin Classifier) : soft margin을 이용해 일부 training data의 잘못된 구분을 허용하는 classifier

Regularization parameter C
* C가 클 경우 : 클래스가 잘못 분류되는 것에 대한 허용치가 크므로 margin이 넓다.
* C가 감소할 경우 : 잘못 분류되는 것에 대한 허용치가 작아지므로 margin이 좁아진다.

Linear boundary가 유효하지 않은 경우
* C와는 무관하게 linear boundary가 클래스 구분에 효과적이지 않을 수 있다.
* Non-linearity를 반영하기 위해서는 *feature의 확장*이나 *kernel 사용*이 필요하다.

**Support Vector Machines**

Feature 확장 (expansion)
* feature들의 변형을 사용해 feature space를 확장시킨다.
    * ==> feature space가 p차원에서 M차원으로 확장된다.
* 확장된 feature space에서 Support Vector Classifier를 fitting한다.
* 확장된 공간에서의 linear boundary는 본래 feature space에서 non-linear decision boundary가 된다.

Inner product
* Kernel : 두 벡터 간의 유사성을 정량화

Support Vector Machine
* Non-linear kernel을 이용한 support vector classifier ex) polynomial kernel (다항식 커널), radial kernel (가우시안 커널)
* kernel 방법의 장점 (feature 확장 대비)
    * 계산량이 적음 : training data 쌍에 대한 계산
    * radial kernel의 경우, feature space 관점으로 보면 infinite-dimensional (무한 확장 커널 가)에 해당
 
Multi-class 문제에 대한 SVM 적용 방법
1. OVA (One Versus All)
* K개의 '클래스 k와 나머지로 분류하는 two-class SVM classifier' 생성 ==> 가장 큰 𝑓መ𝑘(𝑥∗)의 k로 분류

2. OVO(One Versus One)
* 모두 [[K, 2]]쌍의 '클래스 k와 l로 분류하는 two-class SVM classifier' 생성 ==> 가장 많이 선택된 k로 분류

# MLP - neural networks

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# MNIST 데이터 불러오기
num_classes = 10
input_shape = (28, 28, 1)

# 데이터 불러오기
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# 60000장의 이미지
X_train.shape, y_train.shape
```

```python
# 데이터를 이미지로 표현하기
plt.imshow(X_train[1].reshape(28, 28))
plt.gray()
plt.show()
```

```python
# 이미지 scaling하기 [0 ~ 255] --> [0 ~ 1]
# 정수 데이터 타입 >> 실수 데이터 타입
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255

# 이미지를 (28, 28, 1) 모양으로 변환하기
# 차원 확장
X_train = np.expand_dims(X_train, -1)
# training data 차원 확인
print(X_train.shape)

X_test = np.expand_dims(X_test, -1)

# one-hot encoding 변환하기 --> to_categorical()
y_train = keras.utils.to_categorical(y_train, num_classes)
```

```python
# 모델 구성하기
model = keras.Sequential(
    [
        # 특징 추출하기
        keras.Input(shape=input_shape),
        # convolve
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu'),
        # pooling
        layers.MaxPooling2D(pool_size=(2, 2)),
        # convolve
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
        # pooling
        layers.MaxPooling2D(pool_size=(2, 2)),
        # 2차원 >> 1차원
        layers.Flatten(),
        layers.Dense(num_classes, activation='softmax')
    ]
)

model.summary()
```

```python
# mini-batch size
# 한 번 weight 업데이트 하기 위해 필요한 데이터양
batch_size = 128

# epoch 에폭 (학습 과정에서 전체 데이터 사용 횟수) 반복
# 보통 에폭은 100 정도로 잡는다. (하지만, 시간이 많이 걸림.)
epochs = 15

# 모델 훈련
# compile : 컴퓨터가 알아들을 수 있도록 표현 변경
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# fitting
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
# one-hot encoding 변환하기 --> to_categorical()
y_test = keras.utils.to_categorical(y_test, num_classes)

# 평가하기
score = model.evaluate(X_test, y_test, verbose=0)

print("Test data's loss :", round(score[0]*100, 2))
print("Test data's accuracy :", round(score[1]*100, 2))
```

# PCA clustering

**Principal Components Analysis**

비지도학습의 기본사항
* Y(output, label)이 존재하지 않음.
* 머신러닝의 목표가 지도학습에 비해 불명확하다. --> 차원축소, 군집화
* 학습의 결과에 대해 평가하기 어려움.
* 지도 학습의 전처리 과정으로 유용하다.

비지도학습의 목표(**패턴 분석**)
* 목표 : 데이터에서 흥미로운 패턴을 찾는 것
    * 데이터가 가진 정보를 잘 표현하도록 시각화 (visualization)
    * 데이터에서 subgroup을 식별 (데이터가 어떻게 그룹화가 될 것인가?)
    * 데이터의 차원을 축소
* 2가지 대표적인 비지도학습 방법
    1. 주성분 분석 (PCA)
    * 데이터 시각화 또는 지도학습을 위한 데이터 전처리 (차원 축소) 방법
    
    2. 군집화 (clustering)
    * 데이터에서 알려지지 않은 subgroup을 밝혀내는 다양한 방법
* 비지도학습의 활용
    * 다양한 분야에서 중요성이 매우 커지고 있다.
    * label이 없는 데이터를 수집하는 것이 더 쉽고 labeling에 많은 비용이 소요된다.
    * 비지도학습 자체가 데이터를 이해하는데 도움이 되어 EDA에 활용된다.

**PCA**
* 데이터의 저차원 표현을 만들어내는 방법
    * 원래 feature의 linear conbination(선형 결합)을 찾아냄
    * 찾아낸 combination은 최대의 분산(variance, 변동량)을 가진 것부터 순서대로 나열되며 서로 상관관계가 없음 ==> PC1, PC2의 상관계수 = 0 (두 변수는 독립적이다.)
* PCA의 두 가지 목적
    * 데이터 시각화를 위한 도구
    * 지도학습을 위한 새로운 feature의 생성
* feature의 첫 번째 PC는 feature의 normalized linear combination(정규화된 선형결합)이다.
* 첫 번째 PC의 loading : 원래 feature 앞에 곱해지는 값
    * loading의 크기를 제한하지 않으면 loading이 커짐에 따라 variance가 커진다.

PC 계산
1. n * p 크기의 행렬 형태의 데이터 X를 준비하고 X의 모든 컬럼들은 평균이 0이 되도록 조정
2. 다음 형태의 feature 값들의 linear combination을 찾음.
3. 데이터 행렬 X의 SVD를 통해 최적화 문제를 풀 수 있다. (SVD : 선형대수에서 행렬을 factorization하는 방법)

PC loading vector와 PC score
* PC score : feature를 𝜙1 방향으로 투사한 값

첫 번째 PC 이후의 PC 계산
* 두 번째 PC는 variance가 가장 큰 방향이면 𝑍1과는 상관관계가 없는 방향이다. (수직 방향)

PVE (Proportion of Variance Explained, 설명된 분산의 비율)
* 각각의 PC의 크기와 영향성을 측정하기 위해 PVE를 계산
* m번째 PC의 PVE는 0과 1사이의 값으로 계산된다.
    * ==> 따라서, 모든 PVE의 합은 1이 되고, 누적 PVE 값이 중요하게 고려된다.

# KMeans

**K-means clustering**

클러스터링의 기본사항
* 서브그룹이나 클러스터로 불리는 비슷한 특성을 가진 그룹을 찾는 일반적인 방법
* 서로 비슷한 데이터들이 같은 그룹에 포함되도록 학습
    * 데이터가 비슷하다 또는 다르다 라는 기준과 개념을 명확히 해야한다.
    * 이 기준은 애플리케이션과 데이터에 따라 달라진다.

대표적인 클러스터링 방법
* K-means clustering : 클러스터의 중심(centroid)을 기준으로 계산해, 데이터에서 미리 정해진 수 만큼의 클러스터를 찾는 방법
* Hierarchical clustering : 가장 가까운 두 클러스터를 묶어가며 계층을 형성해 미리 정하지 않은 수 만큼의 클러스터를 찾는 방법 / 결과물 : dendrogram

K-means clustering의 기본 사항
* 각 클러스터에 포함되는 데이터의 index를 가진 집합으로 정의
    * 집합의 속성
    1. 각 데이터는 K개의 클러스터 중 최소 하나에 포함된다.
    2. 클러스터는 겹치지 않는다.

K-means clustering의 세부 사항
* 좋은 클러스터링 : within-cluster variation이 가능한 한 작은 것
* within-cluster variation 측정 : 같은 클러스터 내의 각각의 데이터가 **얼마나 다른지 (떨어져있는지)** 측정하는 방법 사용
* K개의 클러스터로 분리할 때, K개의 클러스터의 **within-cluster variation의 총합이 최소**가 되도록 한다.
* Euclidean distance를 주로 사용함.

K-means clustering 알고리즘
* local optimum을 구하는 간단한 알고리즘을 사용한다.
    * ==> 첫 단계에서 데이터를 임의로 할당하기 때문에 매번 결과가 달라질 수 있다.
* 과정
1. 클러스터링 수행 전 데이터
2. 데이터를 3개의 클러스터에 임의로 할당
3. 클래스 중심 계산, 클러스터가 임의로 할당되어 중심이 겹친다.
4. 각 데이터를 가장 가까운 클래스 중심에 할당한다.
5. 3번 과정을 반복
6. 이 과정을 10번 반복한 후의 결과
* 속성 : objective를 최소화하도록 보증된다.
* 기본사항
    * 미리 클러스터의 수 K를 정하고 클러스터링 수행함. (Hierarchical clustering은 미리 클러스터의 수를 정하지 않는다.)
    * Hierarchical clustering 중 일반적으로 **bottom-up** 또는 agglomerative 타입의 클러스터링 방법이 많이 쓰인다.

# AutoEncoder

**AutoEncoder**

구조
* Input과 Output의 값이 같아지도록 학습하는 뉴럴 네트워크
* Encoder : h = f(x), input을 표현하는 압축된 code를 생성
* Decoder : r = g(h), code로부터 input을 복원

유용성과 학습방법
* output을 input과 완전히 동일하게 복원하는 뉴럴네트워크는 유용하지 않다.
    * ==> 오히려 **완전히 동일하게 복원되지 않도록** 학습시키는 것이 중요하다.
    * *제한조건*을 두어 input에 근사하게 복원되도록 학습시킨다.
        * ==> 데이터의 *중요한 속성만* 학습하도록 한다.
* 차원 축소를 위해 사용한다.
    * PCA와 유사하지만 *비선형성*을 반영할 수 있기 때문에 더 유용하다.
* Multilayer Neural Network의 특별한 경우
    * ==> Mini-batch, gradient descent, back-propagation 등을 사용함.
 
**Generative Adversarial Network (GAN)**
* 생성자가 상대방을 반드시 이겨야 하는 게임이론 시나리오에 기반한 생성 네트워크
    * generator : training data와 유사한 샘플을 생성
    * discriminator : 실제일 확률 (= generator가 생성해낸 가짜가 아닐 확률)
* 학습 과정 동안, generator는 점차 실제와 유사한 데이터를 생성하고 discriminator는 real과 fake 데이터를 더 잘 구별하게 된다.


구조
* Generator, Discriminator는 모두 뉴럴 네크워크이다.
* Generator의 input은 임의의(random) 노이즈(noise)이다.

학습
* discriminator - 실제 데이터와 generator가 생성한 데이터를 입력받아 이를 더 잘 구분하는 방향으로 학습된다.
* generator - generator가 만든 fake 데이터를 discriminator가 구분해냈을 때, penalty를 받아 이를 줄이는 방향으로 학습된다.

**준지도학습 Semi-supervised learning**
* 지도학습과 비지도학습의 두 가지를 조합하는 방법 --> 두 방법 중 하나를 이용해 다른 하나의 성능을 향상시키는 것을 목적으로 한다.
* Semi-supervised learning은 *분류 문제*에 적용되는 경우가 많다.

가정사항
1. Smoothness 가정 : 두 값 x, x'가 input space에서 가깝게 있으면 해당하는 y, y'은 같은 label을 가진다.
2. Low-density 가정 : decision boundary는 데이터의 밀도가 낮은 지역을 지난다.
3. Manifold 가정 : 고차원의 데이터는 보통 저차원의 substructure에 집중되어 있음.

종류
* Inductive (귀납적) 방법
    * 지도학습처럼 모델을 만들고 예측하는 방법
    * Unlabeled 데이터는 모델을 만드는 과정에서 성능을 높이는 데 쓰인다.
* Transductive (변환적) 방법
    * 모델을 만들고 예측하는 것이 아닌 *데이터의 연결성*에 기반한 방법
    * ex) 데이터의 관계성을 그래프로 정의하면 비슷한 데이터가 근접해 연결된다.

예제
* Wrapper method
    * labeled 데이터를 이용해 여러 개의 지도학습 모델을 만들고 이 모델을 이용해 unlabeled 데이터의 label을 예측 (pseudo-labeled 데이터)
    * Pseudo-labeled 데이터까지 포함해 새로운 지도학습 모델을 만드는 과정을 반복한다.
* Unsupervised pre-processing
    * unlabeled 데이터로부터 새로운 feature를 만들어내는 방법
    * PCA나 autoencoder를 이용해 unlabeled 데이터로부터 새로운 feature를 만들고 이를 지도학습 모델을 만드는데 사용한다.

```python
import keras
from keras import layers
from keras.datasets import mnist
```

```python
# 데이터 준비
# 이미지 데이터 !
(X_train, _), (X_test, _) = mnist.load_data()

# 0과 1 사이로 정규화
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255

# 훈련용 이미지, 평가용 이미지 개수 확인하기
print(len(X_train), len(X_test))

# 28 * 28 크기의 이미지 60000장
print(X_train.shape)

# 차원 축소
X_train = X_train.reshape((len(X_train), np.prod(X_train.shape[1:])))
X_test = X_test.reshape((len(X_test), np.prod(X_test.shape[1:])))

# 모델 생성
# 인코딩 되는 차원의 수
encoding_dim = 32

# input layer
input_img = keras.Input(shape=(784, )) # 784 >> 1차원으로 들어감.

# hidden layer (784 차원 >> 32 차원)
encoded = layers.Dense(encoding_dim, activation='relu')(input_img) # Dense() : 1차원 (압축)

# output layer (32차원 >> 784 차원)
decoded = layers.Dense(784, activation='sigmoid')(encoded)

# autoencoder model
autoencoder = keras.Model(input_img, decoded)

# summary
autoencoder.summary()
```

```python
# autoencoder fitting
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

autoencoder.fit(X_train, X_train, epochs=20, batch_size=256, shuffle=True, validation_data=(X_test, X_test))

# 예측값 확인하기
decoded_img = autoencoder.predict(X_test)
```

```python
# 시각화
n = 10

plt.figure(figsize=(20, 4))

for i in range(n):
    # 원본이미지 보여주기
    ax = plt.subplot(2, n, i+1)
    # 2차원 이미지로 복원하기
    plt.imshow(X_test[i].reshape(28, 28))
    # 흑백
    plt.gray()
    # 축이 보이지 않도록 하기
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # 복원이미지 보여주기
    ax = plt.subplot(2, n, i+1+n)
    # 2차원 이미지로 복원하기
    plt.imshow(decoded_img[i].reshape(28, 28))
    # 흑백
    plt.gray()
    # 축이 보이지 않도록 하기
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
plt.show()
```

**Principal Components Regression**

PC를 이용한 회귀
* PCA의 feature를 이용하는 것이 원해 feature를 이용하는 것보다 좋은 성능을 낼 수 있다.

```python
from sklearn.decomposition import PCA
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
raw = pd.read_csv(path+'Advertising.csv', index_col=0)
ad = raw.copy()
print(ad.shape)
ad.head(2)
```

```python
# 데이터 준비
X = ad.iloc[:, :-1]
y = ad[['Sales']]

# 주성분 분석 수행하기
model = PCA(n_components=2)

# fitting
model.fit(X)

# 주성분 비율 확인
print(model.explained_variance_ratio_)

# PC score
z = model.fit_transform(X)

# PCA 한 후, sclaed값 활용하기
from sklearn.model_selection import train_test_split

z_train, z_test, y_train, y_test = \
train_test_split(z, y, test_size=0.2, random_state=42)

# 선형회귀 모델 생성
model = linear_model.LinearRegression()

# fitting
model.fit(z_train, y_train)

# train 예측
z_train_pred = model.predict(z_train)

# test 예측
z_test_pred = model.predict(z_test)

# coefficients
print('coef :', model.coef_) # [pc1, pc2]
# train / test MSE
print('train MSE :', round(mean_squared_error(y_train, z_train_pred), 3))
print('test MSE :', round(mean_squared_error(y_test, z_test_pred), 3))
# 결정계수
print('train R2 score :', round(r2_score(y_train, z_train_pred), 3))
print('test R2 score :', round(r2_score(y_test, z_test_pred), 3))
```
