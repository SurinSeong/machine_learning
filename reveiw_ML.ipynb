# LinearRegression

단순 선형 회귀 - y = ax + b

* **Error의 척도(MSE, MAE 등 - Cost Function)를 낮추기 위해** a, b의 값을 최적화하는 방법을 이용
* Error = |예측값 - 실제값|
* 입력 변수의 개수가 늘어나면 Multinomial Linear Regression이 되고
* Sigmoid 함수 안으로 들어가면 Logistic Regression이 된다.

Linear Regression의 전제 (보통, 정규성, 선형성, 등분산성을 확인한 후, 회귀 분석을 진행한다.)

* Normality 정규성 : 잔차의 정규성
    1. QQ plot를 통한 시각화 자료 확인
        * 데이터가 직선 상에 있음 >> 정규성을 따른다.
    2. 히스토그램을 통한 시각화 자료 확인
        * 히스토그램 막대가 정규분포 곡선 상에 있음 >> 정규성을 따른다.
    3. 정량적 검정
        1) Normality test : Shapiro-Wilk test, Kolmogorov-Smirnov test
            * Shapiro-Wilk : 2000개까지 / Kolmogorow-Smirnov : 2000개 이상
        2) Ordinal logistic regression의 proportional odds assumption test : Score test
        3) Cox regression의 proportional hazard assumption test : Schoenfeld residual test
        * p-value > 0.05일 때 >> 정규성을 위반한다고 볼 수는 없다.
        

* Linearity 선형성 : 두 변수 (독립변수-종속변수) 간의 선형 관계
* Homoscedasticity 잔차의 등분산성 : 잔차의 분산은 종속변수의 값에 따라 달라지지 않음.
* Independence 독립성 : 관측 간의 독립

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 선형 회귀 관련 사이킷런
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

import statsmodels.api as sm
```

* 간단한 예제

```python
# 데이터 준비
X = np.array([[0], [1], [2]])
y = np.array([0, 1.2, 1.6])

# model 변수 저장
model = linear_model.LinearRegression()

# model 훈련하기
model.fit(X, y)

# train 예측
y_pred = model.predict(X)

# test 예측
test_pred = model.predict([[1.5]])
test_pred
```

```python
# coefficient
print('coef :', model.coef_)
```

```python
# 시각화
# 실제값
plt.scatter(X, y, color='red')
# 예측값
plt.scatter(1.5, test_pred, color='green')

# 회귀선 그리기
plt.plot(X, y_pred, color='black', linewidth=1)
plt.xlabel('X')
plt.ylabel('y')

plt.grid()
plt.show()
```

* 실데이터 적용하기

```python
path = '../datasets/ML_data/'

raw = pd.read_csv(path+'Advertising.csv', index_col=0)
ad = raw.copy()
print(ad.shape)
ad.head(2)
```

Error의 기준 (Cost Function)

1. MSE
* Error가 정규분포의 특성을 보일 때 적합하다.
* Outlier의 영향을 상대적으로 많이 받는다.
* ML에서 대용량의 데이터를 다룰 때, Gradient Descent 방식을 이용하는 경우가 많음. 이때, 값이 줄어들면서 미분값도 줄어들어 수렴할 수 있다.

2. MAE
* Outlier의 영향을 많이 받지 않는다.
* 기울기가 일정하기 때문에, 최저점으로 수렴이 안 될 가능성이 있음.

3. Huber, Log-cosh
* MAE의 Outlier에 대해 Robust한 특징과 MSE의 Gradient가 최적점 근처에서 줄어듦으로써 더 정말한 값을 찾을 수 있도록 하는 장점을 결합.

4. Quantile
* 예측값의 분산은 항상 일정하다는 것을 바꾸지 않고 quantile을 이용해 예측 구간 표현 가능

```python
# 선형회귀모델 적용하기
# train/test data 분리하기 >> 과적합 방지
train = ad[:-20]
test = ad[-20:]

# train feature / label
X_train = train.iloc[:, :-1]
y_train = train[['Sales']]

# test feature / label
X_test = test.iloc[:, :-1]
y_test = test[['Sales']]
```

* 사이킷런 사용

```python
# TV
# 모델링 - 선형회귀
model = linear_model.LinearRegression()

# 모델 학습
model.fit(X_train[['TV']], y_train)

# train data 예측하기
y_train_pred = model.predict(X_train[['TV']])

# test data 예측하기
y_test_pred = model.predict(X_test[['TV']])

# 기울기
print('coef :', model.coef_)
print('y절편 :', model.intercept_)
# MSE 확인
print("train data's MSE :", round(mean_squared_error(y_train, y_train_pred), 3))
print("test data's MSE :", round(mean_squared_error(y_test, y_test_pred), 3))
# 결정계수 확인
print("train data's R^2 score :", round(r2_score(y_train, y_train_pred), 3))
print("test data's R^2 score :", round(r2_score(y_test, y_test_pred), 3))
```

```python
y_test_pred
```

```python
# 시각화하기
# 산점도
plt.scatter(X_train.TV, y_train.Sales, color='gray')
plt.scatter(X_test.TV, y_test.Sales, color='red')
# 회귀선 그리기
plt.plot(X_train.TV, y_train_pred, color='green', linewidth=2)
# 예측값 확인하기
plt.scatter(X_test.TV, y_test_pred, color='blue', marker='^')

# 그래프 꾸미기
plt.title('Linear Regression Result')
plt.xlabel('TV')
plt.ylabel('Sales')

plt.grid()
plt.show()
```

* statsmodels 사용

```python
# OLS 사용
# train / test data에 x0 컬럼 만들어주기
# 왜냐면, OLS 함수는 데이터 내에 intercept하는 feature가 필요하다.
# train
sm_X_train = X_train.copy()
sm_X_train['x0'] = 1

# test
sm_X_test = X_test.copy()
sm_X_test['x0'] = 1

# 모델 생성하고 훈련하기
model = sm.OLS(y_train, sm_X_train[['x0', 'TV']]).fit()

# train data로 예측하기
y_train_pred = model.predict(sm_X_train[['x0', 'TV']])

# test data로 예측하기
y_test_pred = model.predict(sm_X_test[['x0', 'TV']])

# 결과 확인하기
model.summary()
```

```python
y_test_pred
```
