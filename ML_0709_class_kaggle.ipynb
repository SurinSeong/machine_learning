```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 데이터 전처리

```python
# SimpleImputer 사용
# 결측치 대체 >> 평균값 !
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

X_train[['horsepower_fillna']] = imputer.fit_transform(X_train[['horsepower']])
X_test[['horsepower_fillna']] = imputer.fit_transform(X_test[['horsepower']])

X_train.isnull().sum(), X_test.isnull().sum()
```

# 필요한 컬럼 추출하기

```python
print(X_train.columns)
X_train.head(2)
```

```python
# 컬럼 나누기
# 사용하지 않을 컬럼
del_columns = ['name', 'horsepower']
# 숫자형 데이터
col_num = ['mpg', 'cylinders', 'displacement', 'horsepower_fillna', 'weight', 'acceleration', 'model_year']
# 범주형 데이터
col_cat = []
# 종속변수 컬럼
y_col = ['isUSA']
```

# 데이터 모델링

```python
# 분석 데이터셋 준비
from sklearn.model_selection import train_test_split

# 위에서 나눠준 컬럼을 이용해 진짜 사용할 training set 다시 만들어주기
X_train_use = X_train.drop(columns=del_columns)

# training / validation dataset
X_train_new, X_val, y_train_new, y_val = \
train_test_split(X_train_use, y_train, test_size=0.3, random_state=42)

print(X_train_new.shape, y_train_new.shape, X_val.shape, y_val.shape)
```

## 데이터 스케일링

```python
X_train_new.head(2)
```

```python
# 정규화 >> StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train_new[col_num])

X_train_new[col_num] = scaler.transform(X_train_new[col_num])
X_val[col_num] = scaler.transform(X_val[col_num])
```

```python
# 2차원 -> 1차원
y_train_new.values.ravel()
```

```python
# 모델링 - KNN
from sklearn.neighbors import KNeighborsClassifier

# 모델 생성
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
# 모델 fitting
knn.fit(X_train_new, y_train_new.values.ravel())
```

```python
# 모델링 - 의사결정나무
from sklearn.tree import DecisionTreeClassifier

# 모델 생성
dt = DecisionTreeClassifier(max_depth=10, random_state=42)
# 모델 fitting
dt.fit(X_train_new, y_train_new)
```

## 모델 성능 평가

```python
# KNN 예측
y_val_pred_knn = knn.predict(X_val)
y_val_pred_proba_knn = knn.predict_proba(X_val)

# DT 예측
y_val_pred_proba_dt = dt.predict_proba(X_val)
```

```python
y_val.values.ravel(), y_val_pred_knn
```

```python
y_val_pred_proba_knn[:, 1]
```

```python
# ROC-AUC
from sklearn.metrics import roc_auc_score

score_knn = roc_auc_score(y_val, y_val_pred_proba_knn[:, 1])
score_dt = roc_auc_score(y_val, y_val_pred_proba_dt[:, 1])

print(score_knn, score_dt)
```

```python
best_model = None
best_score = 0

for i in range(2, 10):
    model = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
    model.fit(X_train_new, y_train_new.values.ravel())
    y_val_pred_proba = model.predict_proba(X_val)

    score = roc_auc_score(y_val, y_val_pred_proba[:, 1])
    print(f'{i}개의 이웃일 때의 점수 : {score}')

    if best_score <= score:
        best_score = score
        best_model = model
```

```python
print(best_model, best_score)
```

```python
X_test_new = X_test.drop(columns=del_columns)
X_test_new[col_num] = scaler.transform(X_test_new[col_num])

best_model.predict_proba(X_test_new)
```

# Penguin

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 준비
path = '../datasets/ML_data/'

X_train_raw = pd.read_csv(path+'penguin_X_train.csv')
X_test_raw = pd.read_csv(path+'penguin_X_test.csv')
y_train_raw = pd.read_csv(path+'penguin_y_train.csv')

X_train = X_train_raw.copy()
X_test = X_test_raw.copy()
y_train = y_train_raw.copy()

print(X_train.shape, X_test.shape, y_train.shape)
```

```python
X_train.head(2)
```

bill_length : 부리길이, flipper_length : 날개길이

```python
y_train.head(2)
```

body_mass : 펭귄무게(g)
* 무게를 예측하는 문제

```python
X_train.info()
```

```python
X_test.info()
```

X_train : null값 존재함.

```python
X_train.describe(include='all')
```

```python
train_data = pd.concat([X_train, y_train], axis=1)
train_data.head(2)
```

```python
# 결측치가 있는 행 확인하기
train_data.loc[train_data.sex.isnull()]
```

```python
train_data.isnull().sum()
```

예측하려는 y_train 값에 null 값이 있는데 이것은 제거하는것이 좋을 것 같음.

그런데 sex의 null값은 8개인데 이 중 다른 값이 있는 것은 6개이다. 이것들은 예측해서 넣어줄 수도 있을 것 같은 생각이 든다.

그래서 내가 한다면 sex의 null을 확인한 후, 다른 숫자 데이터로 sex를 예측한 후, 임의로 넣어주고 몸무게를 예측하는 모델을 만들 것 같다.

수업에서 한 것은 제거 ! 왜냐면 결측치의 비율이 0.03밖에 되지 않기 때문 !

```python
train_data_dropna = train_data.dropna().reset_index(drop=True)
train_data_dropna.info()
```

```python
# 독립변수, 종속변수 분리하기
X_train_dropna = train_data_dropna.iloc[:, :-1]
y_train_dropna = train_data_dropna.body_mass_g
```

* 숫자형 데이터는 스케일링이 필요하고
* 문자형 데이터는 원핫인코딩이 필요하다는 것을 알게 되었음.
* train, test 둘 다 바꿔주어야 한다.

```python
# train, test 합쳐주기
X = pd.concat([X_train_dropna, X_test])
X.head(2)
```

```python
# 원핫인코딩할 변수명
col_cat = ['species', 'island', 'sex']
col_num = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore')

encoder.fit(X[col_cat])

X_train_dropna_encoded = encoder.transform(X_train_dropna[col_cat])
X_test_encoded = encoder.transform(X_test[col_cat])
```

```python
print(X_train_dropna_encoded) # sparse matrix : 희소행렬
```

```python
# todense() : 밀집행렬 dense matrix
X_train_dropna_encoded.todense()
```

```python
encoder.get_feature_names_out()
```

```python
# 'X_train_dropna' DF와 합쳐주기
X_train_final = pd.concat([pd.DataFrame(X_train_dropna_encoded.todense(), columns=encoder.get_feature_names_out()), X_train_dropna[col_num]], axis=1)
X_test_final = pd.concat([pd.DataFrame(X_test_encoded.todense(), columns=encoder.get_feature_names_out()), X_test[col_num]], axis=1)

print(X_train_final.shape, X_test_final.shape)
```

```python
# 데이터 모형 구축하기
from sklearn.model_selection import train_test_split

X_train_new, X_val, y_train_new, y_val = \
train_test_split(X_train_final, y_train_dropna, test_size=0.3, random_state=42)

print(X_train_new.shape, y_train_new.shape, X_val.shape, y_val.shape)
```

```python
# 숫자형 데이터 스케일링하기
# X_train_new, X_val, X_test_final
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train_new[col_num])

X_train_new[col_num] = scaler.transform(X_train_new[col_num])
X_val[col_num] = scaler.transform(X_val[col_num])
X_test_final[col_num] = scaler.transform(X_test_final[col_num])
```

```python
# 모델 - 선형회귀
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train_new, y_train_new)

y_val_pred = lr.predict(X_val)

print('BIAS :', lr.intercept_)
print('WEIGHT :', lr.coef_)
```

```python
# 가중치 확인
pd.DataFrame(data=lr.coef_, index=X_train_new.columns, columns=['weight']).sort_values(by='weight', ascending=False)
```

```python
# 모델 평가
from sklearn.metrics import mean_squared_error, r2_score

print('MSE : {:.4f}'.format(mean_squared_error(y_val, y_val_pred)))
print('RMSE : {:.4f}'.format(np.sqrt(mean_squared_error(y_val, y_val_pred)))) # mean_squared_error에서 squared=False해주면 루트한 값이 나온다.
print('R2 SCORE : {:.4f}'.format(r2_score(y_val, y_val_pred)))
```

```python
# test의 예측값 확인
y_pred = lr.predict(X_test_final)
y_pred
```

```python
# DF 만들기
test_final = X_test_final.copy()

test_final['body_mass_g'] = y_pred
test_final.to_csv(path+'penguin_test_result.csv', index=False)
```
