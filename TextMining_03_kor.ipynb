```python
import json
import re
import matplotlib
import matplotlib.pyplot as plt

# 한글 품사 태깅을 위한 모듈
from konlpy.tag import Okt
from collections import Counter
from matplotlib import font_manager, rc
from wordcloud import WordCloud

plt.rcParams['font.family'] = 'NanumGothic'
```

## 데이터 준비

```python
inputFileName = '../datasets/ML_data/etnews.kr_facebook_2016-01-01_2018-08-01_4차 산업혁명'

# # 첫 번째 방법
# # json 파일을 읽어서 data에 저장, 한글 utf-8
# data = json.loads(open(inputFileName + '.json', 'r', encoding='utf-8').read())
# data.close()
# data

# 두 번째 방법
with open(inputFileName + '.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

data
```

* 'message'키의 데이터에서 품사가 명사인 단어만 추출
  

```python
message = ''

for item in data:
    if 'message' in item.keys():
        # 'message' 키의 값에서 문자/숫자가 아닌 것은 공백으로 치환, 연결해서 전체를 하나의 문자열로 구성
        message = message + re.sub(r'[^\w]', ' ', item['message']) + ''

message
```

```python
# 품사 태깅 패키지인 Okt를 사용
nlp = Okt()
# 명사만 추출해 message_N에 저장
message_N = nlp.nouns(message)
message_N
```

## 데이터 탐색 및 분석 모델 구축

* 명사를 추출해 저장한 message_N에 있는 단어들을 탐색

```python
# Counter 함수를 이용해 단어별 출현 횟수 계산
count = Counter(message_N)
count
```

```python
# 출현 횟수가 많은 상위 80개의 단어 중에서 길이가 1보다 큰 것만 word_count 딕셔너리에 저장
word_count = dict()

for tag, counts in count.most_common(80):
    if len(str(tag)) > 1:
        word_count[tag] = counts

word_count
```

* 히스토그램을 그려 단어 빈도를 시각적으로 탐색

```python
plt.figure(figsize=(12, 6))
plt.xlabel('keyword')
plt.ylabel('count')
plt.grid(True)

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation=75)
plt.show()
```

## 결과 시각화

```python
font_path = 'c:/Windows/Fonts/NanumGothic.ttf'
wc = WordCloud(font_path, background_color='ivory', width=800, height=600)
cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize=(8, 8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
cloud.to_file(f'{inputFileName}_cloud.jpg')
```
